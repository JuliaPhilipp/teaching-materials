```python
from palmerpenguins import load_penguins
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load example dataset
penguins = load_penguins()
```

```python
# scatterplot
sns.relplot(
    data=penguins,
    x="flipper_length_mm", y="bill_length_mm", hue="species",
    style="species", 
)

# scatterplot w/ LM
sns.lmplot(
    data=penguins,  x="flipper_length_mm", y="bill_length_mm", hue="species", 
)

# faceting
sns.lmplot(
    data=penguins,  x="flipper_length_mm", y="bill_length_mm", hue="species", col="species",
)
```

```python
# different kinds of catplot
sns.catplot(
    data=penguins, kind="swarm", x="species", y="bill_length_mm", hue="species"
    )

# use kind to change subtype
sns.catplot(
    data=penguins, kind="violin", x="species", y="bill_length_mm", hue="species"
    )
```
